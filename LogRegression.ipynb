```python
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

# 3D scatter plot
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

# Extract data from the DataFrame
z = df['test_result']
x = df['age']
y = df['physical_score']

# Create scatter plot
scatter = ax.scatter(x, y, z, c=z, cmap='viridis', s=50, alpha=0.7)

# Axis labels
ax.set_xlabel('Test Result')
ax.set_ylabel('Age')
ax.set_zlabel('Physical Score')
ax.set_title('3D Scatter Plot: Test Result vs Age vs Physical Score')

# Add color bar based on physical_score
fig.colorbar(scatter, ax=ax, label='Physical Score')

plt.show()
```

```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
```

```python
X = df.drop('test_result',axis = 1)
y = df['test_result']
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.1, random_state=101)
```

```python
scaler = StandardScaler()
scaled_X_train = scaler.fit_transform(X_train)
scaled_X_test = scaler.transform(X_test)
```

```python
scaled_X_test
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
log_model = LogisticRegression()
```

```python
log_model.fit(scaled_X_train,y_train)
```

```python
y_pred = log_model.predict(scaled_X_test)
```

```python
y_pred
```

```python
y_pred_proba = log_model.predict_proba(scaled_X_test)
```

```python
y_pred_proba
```
